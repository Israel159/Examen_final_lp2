```python
import pandas as pd

# Cargar los datos desde los archivos CSV
datos1 = pd.read_csv('datos_filtrados1.csv')
datos2 = pd.read_csv('unmsm_tesis.csv')

# Mapear nombres de columnas entre ambos archivos
mapeo_columnas = {
    'dc.contributor.advisor': 'Asesor',
    'dc.contributor.author': 'Nombre de tesista',
    'dc.date.issued': 'Año de publicación',
    'dc.description.abstract[es_PE]': 'Resumen',
    'dc.publisher[es_PE]': 'Institución',
    'dc.title[es_PE]': 'Titulo de la tesis',
    'thesis.degree.name[es_PE]': 'Grado'
}






```

```python
# Renombrar columnas en el primer conjunto de datos
datos1 = datos1.rename(columns=mapeo_columnas)

# Concatenar por filas
datos_combinados = pd.concat([datos1, datos2], ignore_index=True)

# Guardar el resultado en un nuevo archivo CSV
datos_combinados.to_csv('archivo_combinado.csv', index=False)
```

```python

# Mostrar información sobre el DataFrame
informacion_dataFrame = datos_combinados.info()

# Conteo de valores únicos en una columna específica
conteo_valores_unicos = datos_combinados['Año de publicación'].value_counts()

# Filtrar solo columnas numéricas para calcular la correlación
columnas_numericas = datos_combinados.select_dtypes(include=['float64', 'int64'])
correlacion = columnas_numericas.corr()

# Imprimir resultados

print("\nInformación del DataFrame:")
print(informacion_dataFrame)

print("\nConteo de Valores Únicos en 'Año de publicación':")
print(conteo_valores_unicos)

print("\nCorrelación entre Columnas Numéricas:")
print(correlacion)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Agrupar por la columna "Grado"
grupos_por_grado = datos_combinados.groupby('Grado')

# Visualizar algunos aspectos de los grupos
for nombre_grado, grupo in grupos_por_grado:
    print(f"\nDatos para el Grado: {nombre_grado}")
    print(grupo.head())

    # Crear un gráfico de barras para la columna "Año de publicación"
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Año de publicación', data=grupo)
    plt.title(f"Distribución de Años de Publicación para {nombre_grado}")
    plt.show()

```
