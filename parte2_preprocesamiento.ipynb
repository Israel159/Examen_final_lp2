```python
import pandas as pd
import io

# Leemos el archivo desde la memoria
datos = pd.read_csv(io.BytesIO(uploaded['tesis_unalm.csv']))
```

```python
datos.head()
```

```python
#vamos a comenzar a filtrar solo las variables que nos importan

# Seleccionamos las columnas especificas
columnas_seleccionadas = ['dc.contributor.advisor', 'dc.contributor.author','dc.date.issued', 'dc.description.abstract[es_PE]', 'dc.publisher[es_PE]','dc.source[es_PE]','thesis.degree.name[es_PE]']

# Creamos un nuevo DataFrame con las columnas seleccionadas
datos_filtrados = datos[columnas_seleccionadas]

print(datos_filtrados.head())
```

```python
#¿que valores toma la variable thesis.degree.name[es_PE] ?

# Obtiene todos los valores únicos de 'thesis.degree.name[es_PE]'
valores_unicos = datos_filtrados['thesis.degree.name[es_PE]'].unique()

# Muestra la lista de valores únicos
print(valores_unicos)
```

```python
# Filtramos las filas donde 'thesis.degree.name[es_PE]' toma los valores de interes
valores_deseados = ['Ingeniero Estadístico e Informático', 'Magister Scientiae - Estadística Aplicada']
datos_filtrados_nuevos = datos_filtrados[datos_filtrados['thesis.degree.name[es_PE]'].isin(valores_deseados)]

print(datos_filtrados_nuevos.head())
```

```python
#exportamos la base de datos en formato csv
from google.colab import files

ruta_salida_csv = "/content/datos_filtrados.csv"
datos_filtrados_nuevos.to_csv(ruta_salida_csv, index=False)

files.download(ruta_salida_csv)
```
